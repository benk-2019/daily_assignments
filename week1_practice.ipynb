```python
#Python OOP

class ComputerComponent:
    def __init__(self, manuf="N/A", model="N/A", ser_num=0):
        self.manuf = manuf
        self.model = model
        self.ser_num = ser_num

    def __str__ (self):
        return f"Manufacturer: {self.manuf}\nModel: {self.model}\nSerial Number: {self.ser_num}\n"
    
```

```python
comp = ComputerComponent("Corsair", "T100 ATX Case", "787812673")
print(comp)
```

```python
class CPU(ComputerComponent):
    def __init__(self, manuf="N/A", model="N/A", ser_num=0, cores=0, freq = 0.0):
        self.cores = cores
        self.freq = freq
        super().__init__(manuf,model, ser_num)

    def __str__(self):
        return super().__str__() + f"Cores: {self.cores}\nFrequency: {self.freq}Ghz\n"


class Memory(ComputerComponent):
    def __init__(self, manuf="N/A", model="N/A", ser_num=0, capacity=0.0, freq = 0.0, gen="N/A"):
        self.capacity = capacity
        self.freq = freq
        self.gen = gen
        super().__init__(manuf,model, ser_num)

    def __str__(self):
        return super().__str__() + f"Capacity: {self.capacity}MB\nFrequency: {self.freq}Mhz\nGeneration: {self.gen}\n"


class Storage(ComputerComponent):
    def __init__(self, manuf="N/A", model="N/A", ser_num=0, capacity=0, bus="N/A"):
        self.capacity = capacity
        self.bus = bus
        super().__init__(manuf,model, ser_num)

    def __str__(self):
        return super().__str__() + f"Capacity: {self.capacity}GB\nBus: {self.bus}\n"
    
```

```python
cpu = CPU("Arm", "A7", 1289737893, 2, 1.8)
print(cpu)

mem = Memory("Samsung", "CAP_6GB_SAM_SOLD_DDR4", 678931246, 6.0, 2400.0, "DDR4")
print(mem)

stor = Storage("WD", "WD_GREEN_512GB_JLFDS", 10274930989, 512, "Sata 6")
print(stor)
```

```python
class Computer():
    def __init__(self, cpu, mem, storage):
        self.__cpu = cpu
        self.__mem = mem
        self.__storage = storage

    def __str__(self):
        return self.__cpu.__str__() + "\n" + self.__mem.__str__() + "\n" + self.__storage.__str__()
```

```python
computer = Computer(cpu, mem, stor)
print(computer)
```

```python
#Pandas
#Ex1

import pandas as pd

df = pd.read_csv('./daily_assignments_week_1/assingment_pandas/imdb_top_1000.csv')
df
```

```python
def df_better(df):
    return df.drop(axis=1, columns=["Poster_Link", "No_of_Votes", "Certificate", "Overview", "Meta_score", "Runtime"])

df_better(df)
```

```python
df_useful = df_better(df)
df_useful["Released_Year"]=pd.to_numeric(df_useful["Released_Year"], errors='coerce')
df_useful.dropna(axis=0, subset=["Released_Year"], inplace=True)
df_useful["Released_Year"]=df_useful["Released_Year"].astype(int)

oldest_movie = df_useful[df_useful["Released_Year"]==df_useful["Released_Year"].min()]
oldest_movie
```

```python
newest_movie = df_useful[df_useful["Released_Year"] == df_useful["Released_Year"].max()]
newest_movie
```
