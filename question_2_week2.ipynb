```python
import pandas as pd

#1
df = pd.read_csv("./songs_normalize.csv")
print("Number of Records: " + str(len(df)))
print("")

#2
print("Oldest Song: ")
oldest_song = df[df["year"] == df["year"].min()]
display(oldest_song)
print("")

#3
print("Most Popular Song: ")
most_popular = df[df["popularity"] == df["popularity"].max()]
display(most_popular)
print("")

#4
df["genre"] = df["genre"].str.split(",")
df = df.explode("genre")
df["genre"] = df["genre"].str.strip(" ")
bad_idx = df[df["genre"] == "set()"].index.to_list()
df_genre = df.drop(index=bad_idx)
print(f"There are {len(df_genre["genre"].value_counts())} unique genres")
print("")

#5
print("Most popular genre: " + df_genre["genre"].value_counts().keys()[0])
print("")
#6
df = pd.read_csv("./songs_normalize.csv")
print("Most popular artist: " + df["artist"].value_counts().keys()[0])
print("")

#7
artists = df["artist"].to_list()
song_name = df["song"].to_list()
max = 0
max_idx = 0
i=0
max_len_art = ""
for x in song_name:
    length = len(str(x))
    if(length>max):
        max = len(str(x))
        max_len_art = artists[i]
    i+=1
print("Artist with longest song name: " + max_len_art)
print("")

#8
print("Total Duration of all songs: " + str(df["duration_ms"].sum()) + "ms")
print("")
print("")

#9
print("Second most popular explicit song form 2010: ")
df_explicit = df[(df["explicit"] == True) & (df["year"] == 2010)]
index_2nd_pop_exp = df_explicit["popularity"].sort_values(ascending=False).index[1]
display(df_explicit.loc[index_2nd_pop_exp])
print("")

#10
df = pd.read_csv("./songs_normalize.csv")
df.to_parquet("./songs_normalize.parquet")
print("Size difference based on what windows says: " + str(250-144) + "KB")
```
