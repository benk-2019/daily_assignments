```python
#RUN ALL, find lowest grossing movie per director
df_better = df.drop(columns=["Poster_Link","Overview","No_of_Votes", "Certificate"])
df_gr_lw = df_better
#clean up the gross values
df_gr_lw["Gross"] = df_gr_lw["Gross"].str.replace(",","")#will change df_better, but okay bc only getting rid of ,
#Get rid of rows with missing gross value and then reset the index
df_gr_lw = df_gr_lw.dropna(axis=0,subset=["Gross"]).reset_index(drop=True)
#get a list of all directors
s_directors = df_gr_lw["Director"].unique()
#declare list to track where in our cleaned dataframe the lowest grossing film occurs for a given director
low_idx = []
#???Can do list comprehension???

for i in range(len(s_directors)):
    #get director's movies
    df_director = df_gr_lw[df_gr_lw["Director"] == s_directors[i]]
    #get the index values at which the lowest gorssing movie of our director occurs
    idx_low_dir = df_director[df_director["Gross"] == df_director["Gross"].min()].index.to_list()
    #add that to the list of movies where the gross is the worst for the given director
    low_idx += idx_low_dir

#creat a dataframe where we use locate to pull all of our indexes out of our clean list
df_low_gross = df_gr_lw.loc[low_idx]
df_low_gross#call to print, it's cleaner looking
```
