```python
import requests
import json
import xmltodict
import pandas as pd

url_base = 'https://www.animenewsnetwork.com'
url_api = '/encyclopedia'
endpoint = '/reports.xml?id=155&nlist=all'

anime_url = url_base + url_api + endpoint

params ={
    'type' : 'anime'
}

resp = requests.get(anime_url, params)
data = xmltodict.parse(resp.content)
df_anime = pd.DataFrame(data)
dict_anime = df_anime['report']['item']
df_anime_real = pd.DataFrame(dict_anime)
df_anime_real
```
