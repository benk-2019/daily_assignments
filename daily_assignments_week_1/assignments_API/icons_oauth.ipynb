```python
import requests
from requests_oauthlib import OAuth1
import os
from dotenv import load_dotenv
import json
import pandas as pd

load_dotenv()

auth_key = os.environ["auth_key"]
auth_secret = os.environ["auth_secret"]

auth = OAuth1(auth_key, auth_secret)
ep = "https://api.thenounproject.com/v2/icon/1"

resp = requests.get(ep, auth=auth)
pd.DataFrame(resp.json())


```
