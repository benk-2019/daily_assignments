```python
from dotenv import load_dotenv
import os
import requests
import xmltodict
import pandas as pd
import time
import schedule

load_dotenv()

api_key = os.environ['api_key']
print(api_key)
ver = 8
def get_hot_anime():
    params = {
        'client' : api_key,
        'clientver' : int(ver),
        'protover' : int(1),
        'request' : 'hotanime'
    }

    url = f'http://api.anidb.net:9001/httpapi'
    print(url)
    resp = requests.get(url,params)
    print(resp)
    dict_hotanime = xmltodict.parse(resp.content)
    df_bad = pd.DataFrame(dict_hotanime["hotanime"]["anime"])
    s_almost_clean = pd.Series(df_bad["title"]).astype(str).str.replace("{'@xml:lang': 'x-jat', '@type': 'main', '#text': ", "").str.replace("'","" )
    s_clean = s_almost_clean.str.replace("}", "")
    df_bad["title"] = s_clean
    print(time.strftime("%H:%M:%S"))
    display(df_bad)


schedule.every(60).seconds.do(get_hot_anime)

while True:
    schedule.run_pending()
    time.sleep(1)
```
